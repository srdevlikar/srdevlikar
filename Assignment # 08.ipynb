```python
sns.boxplot(sd['Rented Bike Count'])
```

```python
q3 = sd['Rented Bike Count'].quantile(0.75)
q1 = sd['Rented Bike Count'].quantile(0.25)
iqr = q3-q1
lower_limit = q1-(1.5*iqr)
upper_limit = q3+(1.5*iqr)
```

```python
sns.boxplot(sd[(sd['Rented Bike Count']>=lower_limit) & (sd['Rented Bike Count']<=upper_limit)]['Rented Bike Count'])
```
